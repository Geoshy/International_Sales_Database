### **Issues:**
**1. Delete Rows With Nan OrderId Values.**

**2. Change Data Type of OrderDate Column from Object (String) to Datetime Data Type.**

**3. Change Data Type of Columns (CustomerId - OrderId - OrderNumber - ProductId - SupplierId) from Float to Integer Data Type.**

**4. Rename Columns (City.1 - Country.1 - Phone.1) as They Similar to Previous Columns.**

```python
# We Would Create a Copy of The Main Dataframe for Editing, and Still Have Access to The Main Data.

df_copy = df.copy()
```

**(1) Delete Rows With Nan OrderId Values:**

```python
# Explore The Nan Values in Our Data:
df[df["OrderId"].isnull()]
```

```python
# Drop Nan Values in Column (OrderId), as This Column is Important in Statistical Analysis (Mean = Average).
# First Way Using notna() Method:
df_copy = df_copy[df_copy["OrderId"].notna()]
```

```python
# Second Way Using dropna() Method:
df_copy.dropna(subset=["OrderId"], inplace=True)
```

```python
df_copy.info()
# 2155 Rows 
```

**2. Change Data Type of OrderDate Column from Object (String) to Datetime Data Type:**


```python
# We would Change Column OrderDate from String (Object) into DataTime Using pd.to_datetime Function:
df_copy["OrderDate"] = pd.to_datetime(df_copy["OrderDate"])
```

```python
df_copy.info()
```

**3. Change Data Type of Columns (CustomerId - OrderId - OrderNumber - ProductId - SuplierId) from Float to Integer Data Type:**

```python
# First Way, We Can Use Iterable by for Loop to Change the Data Type of This Columns:

change_columns = ["CustomerId", "OrderId", "OrderNumber", "ProductId", "SupplierId"]
for column in list(df_copy.columns):
    if column in change_columns:
        df_copy[column] = df_copy[column].astype(int)
```

```python
df_copy.info()
```

```python
# Second Way, We Can Create a Function and Apply it on Columns to Convert Seprated Columns Into Integer:
# Create User-Defined Function:
def intconvert(x):
    if pd.notna(x):
        return(int(x))
    else:
        return(x)
```

```python
df_copy["OrderId"] = df_copy["OrderId"].apply(intconvert)
df_copy["CustomerId"] = df_copy["CustomerId"].apply(intconvert)
df_copy["OrderNumber"] = df_copy["OrderNumber"].apply(intconvert)
df_copy["ProductId"] = df_copy["ProductId"].apply(intconvert)
df_copy["SupplierId"] = df_copy["SupplierId"].apply(intconvert)
```

```python
df_copy.info()
```

**4. Rename Columns (City.1 - Country.1 - Phone.1) as They Similar to Previous Columns:**

```python
# We Would Rename Columns That Related to Supplier Which Similar to Order Columns:

df_copy.rename(columns={
    "City.1":"SupplierCity",
    "Country.1":"SupplierCountry",
    "Phone.1":"SupplierPhone"
}, inplace=True)
```
